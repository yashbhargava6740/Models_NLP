```python
nltk.download('stopwords')
stop_word = stopwords.words('english')
```

```python
x = df.loc[:,['selected_text']]
x_values = x.values
```

```python
y = df.loc[:, ['sentiment']]
```

```python
import re
def remove_punc(sent):
  return re.sub(r'[^\w\s]', '', sent)
```

```python
x_values_punc_removed = []
for i in x_values:
    for j in i:
        x_values_punc_removed.append(remove_punc(str(j)))
```

```python
x_values_punc_removed
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('punkt')
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatized_sentences = []
for i in x_values_punc_removed:
    words = word_tokenize(i)
    lemmatized_words = []
    for j in words:
      lemmatized_words.append(lemmatizer.lemmatize(j))
    temp = ' '.join(lemmatized_words)
    lemmatized_sentences.append(temp)
```

```python
lemmatized_sentences
```

```python
# nltk.download('punkt')
x_values_no_stop_words = []
for i in lemmatized_sentences:
    words = word_tokenize(i)
    temp = []
    for j in words:
        if j.lower() not in stop_word:
            temp.append(j)
    temp1 = ' '.join(temp)
    x_values_no_stop_words.append(temp1)
```

```python
x_values_no_stop_words
```

```python
x_new = pd.DataFrame({'sentences': x_values_no_stop_words})
```

```python
x_new
```

```python
y
```

```python
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, SpatialDropout1D
from tensorflow.keras.utils import to_categorical
```

```python
y['sentiment'].value_counts()
```

```python
y_new = y
# y_new
```

```python
sentiment_labels = {'negative': 0, 'neutral': 1, 'positive': 2}
y_new['sentiment'] = y_new['sentiment'].map(sentiment_labels)
y_new = to_categorical(y_new['sentiment'])
```

```python
y_new
```

```python
max_features = 1000  # Maximum number of words to keep
max_len = 50  # Maximum length of each sentence
tokenizer = Tokenizer(num_words=max_features, split=' ')
tokenizer.fit_on_texts(x_new['sentences'].values)
X = tokenizer.texts_to_sequences(x_new['sentences'].values)
X = pad_sequences(X, maxlen=max_len)
```

```python
X
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_new, test_size=0.2, random_state=42)
```

```python
embedding_dim = 128
model = Sequential()
model.add(Embedding(max_features, embedding_dim, input_length=X.shape[1]))
model.add(SpatialDropout1D(0.4))
model.add(LSTM(128, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(3, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
batch_size = 32
epochs = 5
model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_test, y_test), verbose=2)
```

```python
loss, accuracy = model.evaluate(X_test, y_test, verbose=1)
print('Test Accuracy:', accuracy)
```

```python
sentiment_labels_rev = {0: 'negative', 1:'neutral', 2:'positive'}
```

```python
my_sentence_pos = "The best movie I have watched ever"
my_sentence_neg = "This is a worst kind of movie i have ever seen"
my_sentence_neu = "This is a normal movie"

sent_pos = tokenizer.texts_to_sequences([my_sentence_pos])
sent_neg = tokenizer.texts_to_sequences([my_sentence_neg])
sent_neu = tokenizer.texts_to_sequences([my_sentence_neu])

padded_pos = pad_sequences(sent_pos, maxlen=max_len)
padded_neg = pad_sequences(sent_neg, maxlen=max_len)
padded_neu = pad_sequences(sent_neu, maxlen=max_len)

prediction = model.predict(padded_pos)
predicted_label = sentiment_labels_rev[prediction.argmax()]
print(f"Your Sentence: '{my_sentence_pos}' --> Predicted Sentiment: {predicted_label}")

prediction = model.predict(padded_neg)
predicted_label = sentiment_labels_rev[prediction.argmax()]
print(f"Your Sentence: '{my_sentence_neg}' --> Predicted Sentiment: {predicted_label}")

prediction = model.predict(padded_neu)
predicted_label = sentiment_labels_rev[prediction.argmax()]
print(f"Your Sentence: '{my_sentence_neu}' --> Predicted Sentiment: {predicted_label}")
```
